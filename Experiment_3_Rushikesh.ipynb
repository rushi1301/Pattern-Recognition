**Experiment-3** : Maximum Likelihood Estimation.

```python
import random
from matplotlib import pyplot as plt

def generate_data(n_samples, p_red):
  """Generates n_samples data points from a Bernoulli distribution with probability p_red."""
  data = []
  for _ in range(n_samples):
    data.append(1 if random.random() < p_red else 0)
  return data

def mle_red_proportion(data):
  """Estimates the proportion of red marbles using MLE."""
  # Calculate the number of red marbles
  n_red = sum(data)

  # Estimate the proportion (MLE)
  p_red_hat = n_red / len(data)
  return p_red_hat


# Generate random data (replace with your desired values)
n_samples = 100
p_red = 0.5  # True proportion of red marbles (unknown)
data = generate_data(n_samples, p_red)
plt.stem(data)

# Estimate the proportion of red marbles using MLE
p_red_hat = mle_red_proportion(data)

# Print the results
print("Number of samples:", n_samples)
print("Estimated proportion of red marbles:", p_red_hat)
```

**Exercise 1:** Perform the MLE experiment for Gaussian data, (here, your generate_data function will generate the values using Gaussian distribution) and verify the mean value of MLE estimates.

```python
import random  
import numpy as np  
from scipy.stats import norm  
import matplotlib.pyplot as plt  

def generate_gaussian_data(n_samples, mu, sigma):  
    """Generates n_samples data points from a Gaussian distribution with mean mu and standard deviation sigma."""  
    data = np.random.normal(mu, sigma, n_samples)  
    return data  

def mle_gaussian_mean(data):  
    """Estimates the mean of a Gaussian distribution using MLE."""  
    # Calculate the mean of the data  
    mu_hat = np.mean(data)  
    return mu_hat  

mu_hats = []  
for _ in range(100):  
    # Generate random data (replace with your desired values)  
    n_samples = 1000  
    mu = 5  # True mean of the Gaussian distribution (unknown)  
    sigma = 1  # True standard deviation of the Gaussian distribution  
    data = generate_gaussian_data(n_samples, mu, sigma)  
    
    # Estimate the mean of the Gaussian distribution using MLE  
    mu_hat = mle_gaussian_mean(data)  
    mu_hats.append(mu_hat)  

# Print the results  
print("Mean Value of MLE Estimates:", np.mean(mu_hats))  
plt.hist(mu_hats, bins=10, alpha=0.7, color='blue', edgecolor='black')  
plt.axvline(x=mu, color='red', linestyle='--')  
plt.legend(["True Mean", "MLE Estimates"])  
plt.xlabel("Mean Values")  
plt.ylabel("Frequency")  
plt.title("Histogram of MLE Estimates for Gaussian Data")  
plt.show()
```

**Exercise 2:** Perform the MLE experiment for Poisson data, (here, your generate_data function will generate the values using Poisson distribution) and verify the mean value of MLE estimates.

```python
import random  
import numpy as np  
import matplotlib.pyplot as plt  
from scipy.stats import poisson  

def generate_poisson_data(n_samples, lam):  
    """Generates n_samples data points from a Poisson distribution with rate lam."""  
    data = np.random.poisson(lam, n_samples)  
    return data  

def mle_poisson_rate(data):  
    """Estimates the rate of a Poisson distribution using MLE."""  
    # Calculate the sample mean of the data  
    lam_hat = np.mean(data)  
    return lam_hat  

lam_hats = []  
for _ in range(100):  
    # Generate random data (replace with your desired values)  
    n_samples = 1000  
    lam = 5  # True rate of the Poisson distribution (unknown)  
    data = generate_poisson_data(n_samples, lam)  
    
    # Estimate the rate of the Poisson distribution using MLE  
    lam_hat = mle_poisson_rate(data)  
    lam_hats.append(lam_hat)  

# Print the results  
print("Mean Value of MLE Estimates:", np.mean(lam_hats))  
plt.hist(lam_hats, bins=10, alpha=0.7, color='blue', edgecolor='black')  
plt.axvline(x=lam, color='red', linestyle='--')  
plt.legend(["True Rate", "MLE Estimates"])  
plt.xlabel("Rate Values")  
plt.ylabel("Frequency")  
plt.title("Histogram of MLE Estimates for Poisson Data")  
plt.show()
```
